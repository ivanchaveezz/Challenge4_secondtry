```python
import pandas as pd
from pathlib import Path
```

```python
schooldata = Path("Resources/schools_complete.csv")
studentdata = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(schooldata)
student_data = pd.read_csv(studentdata)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
school_count= pd.DataFrame(school_data['school_name'])
school_count1=len(school_count.index)
school_count1
```

```python
student_count=pd.DataFrame(student_data['student_name'])
student_count1=len(student_count.index)
student_count1
```

```python
total_budget=school_data['budget']
total_budget1=sum(total_budget)
total_budget1
```

```python
math_average=student_data['math_score']
math_average1=math_average.mean()
math_average1
```

```python
score_reading=student_data['reading_score']
score_reading1=score_reading.mean()
score_reading1
```

```python
passing_math_count=student_data[(student_data["math_score"] >= 70)].count()['student_name']
passing_math_count1=(passing_math_count/len(student_count.index))*100
passing_math_count1
```

```python
passing_reading_count=student_data[(student_data["reading_score"] >= 70)].count()['student_name']
passing_reading_count1=(passing_reading_count/len(student_count.index))*100
passing_reading_count1
```

```python
passing_math_reading_count = student_data[
    (student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  len(student_count.index)* 100
overall_passing_rate
```

```python
columns=['total school','total students','total budget','average math score','average reading score','%passing math','% passing reading','%overall passing']
data=[school_count1,student_count1,total_budget1,math_average1,score_reading1,passing_math_count1,passing_reading_count1,overall_passing_rate]

district_summary = pd.DataFrame(data)
district_summary=district_summary.transpose()
district_summary.columns=columns


#district_summary["total students"] = district_summary["total students"].map("{:,}".format)
#district_summary["total budget"] = district_summary["total budget"].map("${:,.2f}".format)


district_summary
```

# Scores by School Spending

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_spending_df = per_school_summary.copy()
school_spending_df
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].apply(lambda x:x.replace('$',''))
school_spending_df["Per Student Budget"] = pd.to_numeric(school_spending_df["Per Student Budget"])
```

```python
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['Per Student Budget'],spending_bins,right=True)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["math_score"].mean())
spending_reading_scores = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["reading_score"].mean())
spending_passing_math = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing math"].mean())
spending_passing_reading = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing reading"].mean())
overall_passing_spending = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean())
```

```python
spending_summary = pd.merge(spending_math_scores,spending_reading_scores,how="left", on=["Spending Ranges (Per Student)"]).merge(spending_passing_math,how="left", on=["Spending Ranges (Per Student)"]).merge(spending_passing_reading,how="left", on=["Spending Ranges (Per Student)"]).merge(overall_passing_spending,how="left", on=["Spending Ranges (Per Student)"])
spending_summary
```

# Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(school_spending_df['#students'],size_bins,right=True)
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = pd.DataFrame(per_school_summary.groupby(["School Size"])["math_score"].mean())
size_reading_scores = pd.DataFrame(per_school_summary.groupby(["School Size"])["reading_score"].mean())
size_passing_math = pd.DataFrame(per_school_summary.groupby(["School Size"])["% Passing math"].mean())
size_passing_reading = pd.DataFrame(per_school_summary.groupby(["School Size"])["% Passing reading"].mean())
size_overall_passing = pd.DataFrame(per_school_summary.groupby(["School Size"])["% Overall Passing"].mean())
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary =pd.merge(size_math_scores,size_reading_scores,how="left", on=["School Size"]).merge(size_passing_math,how="left", on=["School Size"]).merge(size_passing_reading,how="left", on=["School Size"]).merge(size_overall_passing,how="left", on=["School Size"])

# Display results
size_summary
```
